'작업시작시간','작업종료시간'을 1분 단위로 변경

- Nip

```python
from datetime import datetime

start = datetime.strptime('2022-01-05 11:02:00','%Y-%m-%d %H:%M:%S')
end = datetime.strptime('2022-06-19 16:51:00','%Y-%m-%d %H:%M:%S')

df_time = pd.date_range(start, end, freq='1min')
```

```python
import datetime

df1 = pd.DataFrame(columns=['작업시작시간'])
df1['작업시작시간'] = df_time
df1['작업종료시간'] = df1['작업시작시간'] + datetime.timedelta(minutes=1)
```

```python
df_nipTime = pd.merge(df_nip, df1, left_on = 'Loss 시작', right_on = '작업시작시간', how = 'right')
```

```python
df_nipTime['종료'].fillna(method='ffill', inplace=True)
```

```python
df_nipTime['Loss 시작'] = df_nipTime['작업시작시간']
```

```python
idx = df_nipTime[df_nipTime['Loss 시작']>=df_nipTime['종료']].index

df_nipTime.drop(idx, inplace=True)
df_nipTime.reset_index(drop=True, inplace=True)
```

```python
df_nipTime.drop(['작업시작시간','작업종료시간'], axis=1, inplace=True)
df_nipTime.fillna(method='ffill', inplace=True)
df_nipTime
```

```python
from datetime import datetime
start = datetime.strptime('2022-02-01 14:35:00','%Y-%m-%d %H:%M:%S')
end = datetime.strptime('2022-07-04 14:11:00','%Y-%m-%d %H:%M:%S')

df_time = pd.date_range(start, end, freq='1min')
```

```python
import datetime

df1 = pd.DataFrame(columns=['작업시작시간'])
df1['작업시작시간'] = df_time
df1['작업종료시간'] = df1['작업시작시간'] + datetime.timedelta(minutes=1)
```

```python
df_gwisangTime = pd.merge(df_gwisang, df1, left_on = 'Loss 시작', right_on = '작업시작시간', how = 'right')
```

```python
df_gwisangTime['종료'].fillna(method='ffill', inplace=True)
```

```python
df_gwisangTime['Loss 시작'] = df_gwisangTime['작업시작시간']
```

```python
idx = df_gwisangTime[df_gwisangTime['Loss 시작']>=df_gwisangTime['종료']].index

df_gwisangTime.drop(idx, inplace=True)
df_gwisangTime.reset_index(drop=True, inplace=True)
```
