### 0.데이터준비(중점. 작업시간 중간에 빠진 시간 기록 찾기)

```python
df = pd.read_csv('C:/Users/USER/etri_csv/1/0_result/DataSet.csv')

df.columns
```

```python
df.head()
```

```python
df['작업시작시간'] = pd.to_datetime(df['작업시작시간'])
df['작업종료시간'] = pd.to_datetime(df['작업종료시간'])
```

1. 작업종료시간와 데이터가 같은 작업종료시간_y를 만든다.

```python
df['작업종료시간_y']=""
```

```python
df.columns
```

```python
df = df.loc[:,['설비','가동상태','이력TYPE','작업시작시간','작업종료시간','작업종료시간_y','작업총시간','LOT','제조번호','지종','평량','지폭(규격)','OrderNo','TYPE','릴No','생산량길이(매/속수)','생산량중량','파지량길이','파지량중량','Shift','작업자','속도표준','속도계산','LOTNote','HOLE 정보 M','HOLE 정보 L','HOLE 정보 XL','HOLE 정보 EC','HOLE 땜 정보 M','HOLE 땜 정보 L','HOLE 땜 정보 XL','HOLE 땜 정보 EC']]
#backup : v5
```

```python
df1 = df
```

```python
df1['작업종료시간_y']=df1['작업종료시간']
df1
```

2. 정상적인 기록은 '작업종료시간_y' 값을 삭제한다.

```python
from tqdm import tqdm

idxStart=0

for i in tqdm(range(df1['작업종료시간_y'].count())):
    if df1.loc[idxStart,'작업종료시간_y']>df1.loc[i,'작업종료시간_y']:
        df1.loc[i,'작업종료시간_y']= " "
    elif df1.loc[idxStart,'작업종료시간_y']<df1.loc[i, '작업종료시간_y']:
        idxStart=i
```

```python
df1 #backup : v5
```

```python
df1_1=df1
```

3.df1_1.loc[i,'작업종료시간'] != df1_1.loc[i+1,'작업시작시간'] 일 때

```python
for i in range(df1_1['작업종료시간'].count()):
    if(i<6130):
        print(i, ":",df1_1.loc[i,'작업종료시간'], 'vs ',i+1, ":", df1_1.loc[i+1,'작업시작시간'])
        if df1_1.loc[i,'작업종료시간'] != df1_1.loc[i+1,'작업시작시간']:
            if df1_1.loc[i,'작업종료시간_y'] != " ":
                df1_1.loc[i,'작업종료시간']=df1_1.loc[i+1,'작업시작시간']
            else:
                print(df1.loc[i,'작업종료시간']," ",df1.loc[i+1,'작업시작시간'], '추가')
                new_data = {'설비':' '
                            ,'가동상태':' '
                            ,'이력TYPE':' '
                            ,'작업시작시간': df1.loc[i,'작업종료시간']
                            ,'작업종료시간' : df1.loc[i+1,'작업시작시간']
                            ,'작업종료시간_y':' '
                            ,'작업총시간':' '
                            ,'LOT': ' '
                            ,'제조번호':' '
                            ,'지종':' '
                            ,'평량':' '
                            ,'지폭(규격)':' '
                            ,'OrderNo':' '
                            ,'TYPE':' '
                            ,'릴No':' '
                            ,'생산량길이(매/속수)':' '
                            ,'생산량중량':' '
                            ,'파지량길이':' '
                            ,'파지량중량':' '
                            ,'Shift':' '
                            ,'작업자':' '
                            ,'속도표준':' '
                            ,'속도계산':' '
                            ,'LOTNote':' '
                            ,'HOLE 정보 M':' '
                            ,'HOLE 정보 L':' '
                            ,'HOLE 정보 XL':' '
                            ,'HOLE 정보 EC':' '
                            ,'HOLE 땜 정보 M':' '
                            ,'HOLE 땜 정보 L':' '
                            ,'HOLE 땜 정보 XL':' '
                            ,'HOLE 땜 정보 EC':' '
                           }
                df1_1=df1_1.append(new_data, ignore_index=True) #'ignore_index=True'하지 않으면 이전 데이터에 있던 인덱스를 그대로 가져온다.
            
```

```python
df1_1##backup : v4
```

정렬

```python
temp = df1_1.sort_values(by=['작업시작시간','작업종료시간'], ascending=[True, True])  
temp.reset_index(drop=True, inplace=True)
```

4. 정렬 체크

```python
idx=[]

for i in range(temp['작업종료시간'].count()):
    if(i<6130):
        if temp.loc[i,'작업종료시간'] >temp.loc[i+1,'작업시작시간']:
            idx.append(i)
```

```python
idx
```

4. '설비' 공백 채운다.

```python
temp['설비'].fillna(method='ffill', inplace=True)
```

5. '작업총시간'을 수정한다.

```python
temp.loc[:,'작업총시간'] = (temp.loc[:,'작업종료시간']-temp.loc[:,'작업시작시간'])
temp['작업총시간']=temp['작업총시간'].dt.total_seconds()/60
temp['작업총시간']=temp['작업총시간'].astype(int)
temp
```

6. '작업종료시간_y'컬럼 삭제

```python
temp = temp.drop(['작업종료시간_y'], axis=1)
temp.columns
```

```python
df1 = temp 
df1#backup : v5
```

### 1. 지종&평량마다 누적카운트

```python
df2 = df1
df2['지종&평량_횟수']=np.nan
```

```python
df2_1 = df2[df2['가동상태']=='생산']
```

```python
idx = df2_1.index
df2_1.reset_index(drop=True, inplace=True)
```

```python
df2_1
```

```python
index=0
count=0
i=0

while(i<len(df2_1)+1):
    if i<len(df2_1):
        print(i,":",df2_1.loc[i,"제조번호"], " vs ",index,":",df2_1.loc[index,"제조번호"])
        if df2_1.loc[i,"제조번호"]==df2_1.loc[index,"제조번호"] :
            if df2_1.loc[i,"지종"]==df2_1.loc[index,"지종"] : 
                if df2_1.loc[i,"평량"]==df2_1.loc[index,"평량"]:
                    count=count+1
                    print("count:",count)
            i=i+1
        else:
            print("횟수 저장")
            df2_1.loc[i-1,"지종&평량_횟수"]=count
            count=0
            index=i
    else:
        df2_1.loc[i-1,"지종&평량_횟수"]=count
        break;
```

```python
df2_1
```

```python
df2_1.set_index(keys=[idx], inplace=True, drop=True)
```

```python
df2['지종&평량_횟수']=df2_1.loc[:,"지종&평량_횟수"]
```
