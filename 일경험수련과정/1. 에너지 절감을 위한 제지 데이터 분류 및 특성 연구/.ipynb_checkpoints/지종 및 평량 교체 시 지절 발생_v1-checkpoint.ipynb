```python
import pandas as pd

df_MES = pd.read_csv('C:/Users/USER/etri_csv/1/0_result/DataSet.csv')

df_MES.head(5)
```

```python
df_MES.loc[df_MES['이력TYPE'].isnull() & df_MES['Note'].str.contains('지종교체', na=False), '이력TYPE'] = '지종교체'
df_MES.loc[df_MES['이력TYPE'].isnull() & df_MES['Note'].str.contains('평량교체', na=False), '이력TYPE'] = '평량교체'
```

'지종, 평량, 지폭' 컬럼 공백 처리

```python
df_MES.loc[:,['지종','평량','지폭(규격)']]=df_MES.loc[:,['지종','평량','지폭(규격)']].fillna(method='bfill')
```

이력TYPE 공백 처리

```python
df_MES['이력TYPE'].fillna('', inplace=True)
```

```python
from tqdm import tqdm
import numpy as np

jijong = df_MES.loc[0,'지종']
size = df_MES.loc[0, '평량']
TYPE = df_MES.loc[0, '이력TYPE']


with tqdm(total = len(df_MES)) as pbar:
    i=1
    while i<len(df_MES):
        if (jijong != df_MES.loc[i,'지종']) &(size !=df_MES.loc[i, '평량']):
            if df_MES.loc[i, '이력TYPE']=='':
                df_MES.loc[i, '이력TYPE'] = '지종교체'
            jijong = df_MES.loc[i,'지종']
            size = df_MES.loc[i, '평량']
            TYPE = df_MES.loc[i, '이력TYPE']

        elif (jijong == df_MES.loc[i,'지종']) &(size !=df_MES.loc[i, '평량']):
            if df_MES.loc[i, '이력TYPE']=='':
                df_MES.loc[i, '이력TYPE'] = '평량교체'
            size = df_MES.loc[i, '평량']
            TYPE = df_MES.loc[i, '이력TYPE']

        elif (jijong == df_MES.loc[i,'지종']) &(size ==df_MES.loc[i, '평량']):   
            if df_MES.loc[i, '이력TYPE']==TYPE:
                df_MES.loc[i, '이력TYPE']=np.nan
        pbar.update(1)
        i=i+1
        
```

```python
df_MES.replace( '', np.nan,regex=True,inplace=True)
```
